```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Loading Financial Dataset**



```python
path="/creditcard.csv"
df=pd.read_csv(path)
df.head()
```

**Exploratory Data Analysis**

```python
df.info()
```

```python
fraud = df[df['Class'] == 1]
normal = df[df['Class'] == 0]

print(f"Fraud cases: {len(fraud)}")
print(f"Normal cases: {len(normal)}")
print(f"Fraud %: {len(fraud) / len(df) * 100:.5f}%")
print(f"Normal %: {len(normal) / len(df) * 100:.5f}%")
```

```python
df.describe()
```

```python
print("Dataset shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum().sum())
print("\nClass distribution:\n", df['Class'].value_counts(normalize=True) * 100)
```

**Visualizations**

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(normal['Amount'], bins=50, color='green', label='Normal', stat='density', kde=True)
sns.histplot(fraud['Amount'], bins=50, color='red', label='Fraud', stat='density', kde=True)
plt.legend()
plt.title('Distribution of Transaction Amount')

plt.subplot(1, 2, 2)
sns.histplot(df['Time'], bins=50, kde=True, color='steelblue')
plt.title('Distribution of Transaction Time')
plt.show()
```

```python
plt.figure(figsize=(18, 10))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Feature Correlation Heatmap', fontsize=16)
plt.show()
```

```python
fraud = df[df['Class'] == 1]
normal = df[df['Class'] == 0]

plt.figure(figsize=(8, 5))
sns.countplot(x='Class', data=df, palette='Set3')
plt.title('Fraud vs Normal Class Distribution')
plt.xticks([0, 1], ['Normal (0)', 'Fraud (1)'])
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(fraud['Time'], bins=100, color='green', label='Fraud Transactions', kde=True)
plt.title('Fraud Transactions Over Time')
plt.xlabel('Time (seconds since first transaction)')
plt.legend()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
cor_target = abs(corr['Class']).sort_values(ascending=False)[1:11]

plt.figure(figsize=(8, 5))
sns.barplot(x=cor_target.values, y=cor_target.index, palette='mako')

plt.title('Top 10 Features Most Correlated with Fraud (Class)', fontsize=14, fontweight='bold')
plt.xlabel('Absolute Correlation', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Class', y='V14', data=df, palette='Set2')
plt.title('Boxplot of V14 by Class (Fraud vs Normal)')
plt.xlabel('Class')
plt.ylabel('V14 Value')

plt.subplot(1, 2, 2)
sns.boxplot(x='Class', y='V10', data=df, palette='Set3')
plt.title('Boxplot of V10 by Class (Fraud vs Normal)')
plt.xlabel('Class')
plt.ylabel('V10 Value')

plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Sample 100 fraud and 100 non-fraud transactions
fraud_sample = df[df['Class'] == 1].sample(100, random_state=1)
normal_sample = df[df['Class'] == 0].sample(100, random_state=1)

# Combine them using pd.concat (instead of .append)
sample_df = pd.concat([fraud_sample, normal_sample])

# Plot pairplot on selected features
sns.pairplot(sample_df, vars=['V10', 'V12', 'V14', 'V17'], hue='Class', plot_kws={'alpha': 0.6})
plt.suptitle('Pairplot of Selected Features by Class', y=1.02)
plt.show()

```

**Observations on fraud trends:**


* Fraud amounts tend to cluster in the lower range (< $100).Non-fraudulent(normal) transactions span a wider range, including very
high amounts.
* Fraudulent activity tends to spike during specific time windows.
* Features such as V14, V10, V17, V12, V16 show strong correlation with fraud.
* Pairplots and boxplots of fraud-related features show distinct distribution shapes for fraud vs. normal.Some features have long negative tails only when fraud is present, e.g., V14, V10.
